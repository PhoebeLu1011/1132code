```python
exercise_data = {
    "Cardio": {
        "Running": 300,
        "Cycling": 250,
        "Jump Rope": 180
    },
    "Strength Training": {
        "Weightlifting": 200,
        "Push-ups": 150,
        "Squats": 170
    },
    "Sports": {
        "Basketball": 400,
        "Soccer": 450,
        "Tennis": 350
    }
}
```

```python
import pandas as pd

df = pd.DataFrame(list(exercise_data.keys()), columns=['Category'])
```

```python
df
```

```python
# 轉換為長格式
data_all = []
for category, activities in exercise_data.items():
    for activity, calories in activities.items():
        data_all.append([category, activity, calories])

# 建立 DataFrame
df_all = pd.DataFrame(data_all, columns=["Category", "Activity", "Calories"])
```

```python
from datetime import datetime

# 記錄完成與未完成的次數
completed_count = 0
not_completed_count = 0

# 讓使用者輸入是否完成每個運動
exercise_done = input("今天有完成規定的運動嗎？(yes/no): ").strip().lower()

if exercise_done == "yes":
        completed_count += 1  # 已完成的運動數 +1
else:
        not_completed_count += 1  # 未完成的運動數 +1
date = datetime.now().strftime("%Y-%m-%d")  # 紀錄日期
diet = input("請輸入今天的飲食內容: ").strip()
notes = input("請輸入今天的心得記錄: ").strip()


# 顯示結果
print("\n✅ 統計結果:")
print(f"完成的運動數: {completed_count}")
print(f"未完成的運動數: {not_completed_count}")
```

```python
data = {
    "日期": [date],
    "運動完成": [exercise_done],
    "飲食內容": [diet],
    "心得": [notes],
    "完成次數": [completed_count],
    "未完成次數": [not_completed_count]
}

df2 = pd.DataFrame(data)
```

```python
print("\n✅ 今日記錄:")
print(df2)
```
