```python
exercise_calories = {
    "cardio": {
        "Running": 400,
        "Cycling": 350
    },
    "strength training": {
        "Weightlifting": 200,
        "Push-ups": 100
    },
    "sports": {
        "Basketball": 300,
        "Soccer": 320
    }
}
```

```python
exercise_calories
```

```python
import pandas as pd

df = pd.DataFrame(list(exercise_calories.keys()), columns=['Category'])
```

```python
df
```

```python
# 轉換為長格式
data_all = []
for category, activities in exercise_calories.items():
    for activity, calories in activities.items():
        data_all.append([category, activity, calories])

# 建立 DataFrame
df_all = pd.DataFrame(data_all, columns=["Category", "Activity", "Calories"])
```

```python
df_all
```

```python
from gspread_dataframe import set_with_dataframe

# 選擇要更新的工作表（選擇第一個工作表）
worksheet = gsheets.get_worksheet(0)  # 如果是其他工作表，可以用 get_worksheet(index)

# 將 DataFrame 寫入 Google Sheets
set_with_dataframe(worksheet, df_all)
```

```python
worksheet_weather = gsheets.get_worksheet(1)
```

```python
worksheet_weather
```

```python
# 取得 Google Sheets 內容
data = worksheet_weather.get_all_values()  # 取得所有儲存格資料，回傳為 2D 陣列

# 將資料轉換為 DataFrame
df_weather = pd.DataFrame(data)

# 設定第一行為標題（欄位名稱）
df_weather.columns = df_weather.iloc[0]  # 第一行設為欄位名稱
df_weather = df_weather[1:]  # 刪除第一行，保留數據部分

# 重設索引
df_weather.reset_index(drop=True, inplace=True)
```

```python
df_weather
```

```python
df_part = df_weather[df_weather['日期'] == '2025/01/19']
```
