```python
import google.generativeai as genai
from google.colab import userdata
import random
import time


apikey = userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=apikey)
model = genai.GenerativeModel("gemini-1.5-flash")

def ai_vs_ai_dice_game(rounds=3):
    ai1_wins = 0
    ai2_wins = 0

    for round_number in range(1, rounds + 1):
        print(f"\n第 {round_number} 輪：")
        time.sleep(1)

        ai1_roll = random.randint(1, 6)
        ai2_roll = random.randint(1, 6)

        print(f"AI 1 擲出了：{ai1_roll}")
        print(f"AI 2 擲出了：{ai2_roll}")

        if ai1_roll > ai2_roll:
            print("結果：AI 1 獲勝！")
            ai1_wins += 1
        elif ai2_roll > ai1_roll:
            print("結果：AI 2 獲勝！")
            ai2_wins += 1
        else:
            print("結果：平手！")

        # AI 之間的對話
        prompt = (
            f"AI 1 擲出了 {ai1_roll}，AI 2 擲出了 {ai2_roll}。\n"
            f"請模擬 AI 1 和 AI 2 討論這場比賽的戰略，但討論時可以互相欺騙。"
        )
        try:
            response = model.generate_content(prompt)
            print(f"AI 之間的對話：\n{response.text.strip()}\n")
        except Exception as e:
            response = model.generate_content(prompt)
            print(f"AI 之間的對話：\n{response.text.strip()}\n")
    print("\n遊戲結束！")
    print(f"AI 1 總共贏了 {ai1_wins} 次，AI 2 總共贏了 {ai2_wins} 次")

# 執行遊戲，僅 3 輪
ai_vs_ai_dice_game(rounds=3)
```
